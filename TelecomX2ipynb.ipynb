```python
datos.drop(columns=['customerID', 'Charges.Total'], inplace=True)
```

##**Encoding**##

```python
datos['gender'] = datos['gender'].map({'Female': 0, 'Male': 1})
```

```python
# Lista de columnas categóricas para one-hot encoding (excluyendo 'gender' y 'Churn')
categorical_cols = [
    'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'
]

# Aplicar one-hot encoding
datos_encoded = pd.get_dummies(datos, columns=categorical_cols, drop_first=True)

# Convertir 'Churn' a 0/1 (si quieres que sea numérica)
datos_encoded['Churn'] = datos_encoded['Churn'].map({'No': 0, 'Yes': 1})

print(datos_encoded.head())
```

##**Verificación de la Proporción de Cancelación (Churn)**##

```python
# Contar valores únicos de 'Churn'
conteo = datos['Churn'].value_counts()

# Mostrar conteos
print(conteo)

# Calcular proporciones en %
proporcion = datos['Churn'].value_counts(normalize=True) * 100
print(proporcion)

# Mostrar proporción cancelados vs activos
print(f"Proporción de clientes que cancelaron (Yes): {proporcion.get('Yes', 0):.2f}%")
print(f"Proporción de clientes que permanecieron (No): {proporcion.get('No', 0):.2f}%")
```

##**Balanceo de Clases**##

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Separar variables predictoras y target
X = datos_encoded.drop('Churn', axis=1)
y = datos_encoded['Churn']

print('Distribución original:', Counter(y))

# Crear objeto SMOTE
smote = SMOTE(random_state=42)

# Aplicar SMOTE para balancear
X_resampled, y_resampled = smote.fit_resample(X, y)

print('Distribución después de SMOTE:', Counter(y_resampled))
```

##**Análisis de Correlación**##

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Unimos X_resampled con y_resampled para poder analizar 'Churn'
df_corr = X_resampled.copy()
df_corr['Churn'] = y_resampled

# Seleccionamos solo las columnas numéricas
numeric_cols = df_corr.select_dtypes(include=['int64', 'float64'])

# Calculamos la matriz de correlación
correlation_matrix = numeric_cols.corr()

# Visualizamos la matriz de correlación con un mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Matriz de Correlación')
plt.show()
```

##**Analisis Dirigido**##

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asegúrate de tener la variable 'TotalCharges'
datos_encoded['TotalCharges'] = datos_encoded['Charges.Monthly'] * datos_encoded['tenure']

# Boxplot: Tipo de contrato vs Cancelación
plt.figure(figsize=(8, 5))
sns.countplot(data=datos, x='Contract', hue='Churn')
plt.title('Cancelación según tipo de contrato')
plt.ylabel('Cantidad de clientes')
plt.xlabel('Tipo de contrato')
plt.show()

# Boxplot: TotalCharges vs Cancelación
plt.figure(figsize=(8, 5))
sns.boxplot(data=datos_encoded, x='Churn', y='TotalCharges')
plt.title('Gasto total vs Cancelación')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Total Charges')
plt.show()
```

##**Modelo Predictivo Separación de Datos**##

```python
from sklearn.model_selection import train_test_split

# Dividir en entrenamiento y prueba (80% / 20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled,
    test_size=0.2,       # 20% para prueba
    random_state=42,     # para reproducibilidad
    stratify=y_resampled # asegura que ambas clases estén balanceadas en ambos sets
)

# Confirmar tamaños
print("Tamaño del set de entrenamiento:", X_train.shape)
print("Tamaño del set de prueba:", X_test.shape)
```

##**Creación de Modelos**##

##**Modelo KNN**##

**Normalizando**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Entrenar el modelo KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

# Crear el modelo con k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
```

**Evaluacion**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = knn.predict(X_test_scaled)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Matriz de confusión
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Reporte de clasificación
print("Classification Report:\n", classification_report(y_test, y_pred))
```

##**Modelo Árbol de Decisión**##

**1. Dividir el conjunto de datos**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)
```

**2. Entrenar el Árbol de Decisión**

```python
from sklearn.tree import DecisionTreeClassifier

# Crear y entrenar el modelo
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
```

**3. Evaluar el rendimiento del modelo**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = dt_model.predict(X_test)

# Métricas
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

**Visualizar el árbol**

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plot_tree(dt_model, filled=True, feature_names=X_train.columns, class_names=["No", "Yes"], max_depth=3)
plt.show()
```

Para una mejora real, puedes probar un Random Forest, que suele superar a ambos al reducir overfitting y mejorar precisión.

#*💡 Estrategias de Retención*#
Fomentar Contratos a Largo Plazo Los clientes con contratos mensuales tienen una mayor tasa de cancelación.
Estrategia: ofrecer descuentos o beneficios exclusivos por cambiar a contratos anuales o bianuales.

Atención Temprana a Nuevos Clientes La variable tenure indica que los nuevos clientes cancelan con mayor frecuencia.
Estrategia: implementar campañas de seguimiento durante los primeros meses, con encuestas de satisfacción y soporte personalizado.

Revisar Métodos de Pago Métodos como Electronic Check se relacionan con mayor churn.
Estrategia: incentivar el uso de pagos automáticos con tarjeta o débito directo, ofreciendo beneficios o puntos de fidelización.

Soporte Técnico y Servicio de Internet La falta de TechSupport o el uso de servicios de InternetService de tipo Fiber Optic (posiblemente por problemas técnicos o precio) está asociada con cancelaciones.
Estrategia: monitorear la calidad del servicio de internet, mejorar la atención técnica y ofrecer paquetes combinados que incluyan soporte.
